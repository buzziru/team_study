```python
import pandas as pd
```

```python
iris_pd = pd.DataFrame(iris.data, columns=iris['feature_names'])
iris_pd.head()
```

```python
iris_pd['species'] = iris.target
iris_pd.head()
```

## Decision Tree

```python
plt.figure(figsize=(12, 6))
sns.scatterplot(data=iris_pd, x='petal length (cm)', y='petal width (cm)', hue='species', palette='Set2')
```

```python
iris_12 = iris_pd[iris_pd['species']!=0]
iris_12.info()
```

```python
plt.figure(figsize=(12, 6))
sns.scatterplot(data=iris_12, x='petal length (cm)', y='petal width (cm)', hue='species', palette='Set2')
```

```python
from sklearn.tree import DecisionTreeClassifier

iris_tree = DecisionTreeClassifier()
iris_tree.fit(iris['data'][:, 2:], iris['target'])
```

```python
from sklearn.metrics import accuracy_score

y_pred_tr = iris_tree.predict(iris['data'][:, 2:])
accuracy_score(iris['target'], y_pred_tr)
```
